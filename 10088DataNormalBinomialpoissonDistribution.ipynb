```python
#numpy array for random module for data distribution
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([1,2,4,5,7,8,9], hist=False)
plt.show()

# random normal distribution
x = random.normal(loc=1, scale = 2, size=(2,3))
sns.distplot(x, hist= False)
plt.show()

# random binomial distribution
y = random.binomial(n = 10, p =0.5, size = 10)
sns.distplot(y, hist =False)
plt.show()

# random poisson distribution
z = random.poisson(lam = 2, size = 10)
sns.distplot(z, kde =False)
plt.show()
```
